```python
#Create the dataset
data ={
    'Hours_Studied': [2,4,6,8,10],
    'Classess_Attended': [3,4,5,6,8],
    'Final_Score': [665,70,75,85,95]
}

# Use Ordinary least squares (OLS) model
#Predict final score for a new student who studied 7 hours and attended 5 classes
```

```python
import statsmodels.api as sm
# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Define independent (X) and dependent (y) variables
X = df[['Hours_Studied', 'Classess_Attended']]
y = df['Final_Score']

# Add constant (for intercept)
X = sm.add_constant(X)

# Fit Ordinary Least Squares (OLS) model
model = sm.OLS(y, X).fit()

# Display summary of the model
print(model.summary())

# Predict Final Score for a new student (7 hours studied, 5 classes attended)
new_student = pd.DataFrame({'const': [1], 'Hours_Studied': [7], 'Classess_Attended': [5]})
predicted_score = model.predict(new_student)

print("\nPredicted Final Score for a student who studied 7 hours and attended 5 classes:", predicted_score[0])
```
