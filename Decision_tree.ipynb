```python
# 2. Importing dataset
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AGA/Data/Social_Network_Ads.csv')
dataset = data
dataset
```

```python
# Split Independent Variable  (Column - Split)
X = data.iloc[:, 2:4].values # or data.iloc[:,0:3]
print(X)
```

```python
# Split Dependent Variable
Y = data.iloc[:, -1].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_test
```

```python
# fitting decision tree classification to the training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, Y_train)
```

```python
#pridicting the test set results
Y_pred = classifier.predict(X_test)
Y_pred
```

```python
Y_test
```

```python
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
cm
```

```python
# formula way to find accuracy
Accuracy = (cm[0][0]+cm[1][1])/(cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1])
Accuracy
```

```python
#formula way to find precision
Precision = cm[1][1]/(cm[1][1]+cm[0][1])
Precision
```

```python
# formula way to find recall
Recall = cm[1][1]/(cm[1][1]+cm[1][0])
Recall
```

```python
# formula way to find the f1 score
F1_score = 2*Precision*Recall/(Precision+Recall)
F1_score
```

```python
# formula way  to find support
Support = cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1]
Support
```

```python
# calclate the accuracy from confusion mattrix
from sklearn import metrics
prediction = metrics.accuracy_score(Y_test, Y_pred)
print("Accuracy : ", prediction*100, "%")
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```

```python
from sklearn import tree
tree.plot_tree(classifier)
```
